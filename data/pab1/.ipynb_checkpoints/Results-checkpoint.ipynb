```python
from scipy.stats import pearsonr
from scipy.stats import spearmanr
import matplotlib.pyplot as plt
```

```python
# AAIndex Random Split

pear_AAR_train = [0.91, 0.91, 0.94, 0.95, 0.93]
pear_AAR_test = [0.91, 0.90, 0.93, 0.93, 0.92]
pear_AAR_names = ["Lr_Norm", "Lr", "NN", "CNN", "GCN"]

plt.figure(figsize=(8,5))
plt.ylim(-0.3, 1.0)
plt.grid(color='b', ls = '-.', lw = 0.25, alpha=0.5)
react = plt.bar(pear_AAR_names, pear_AAR_train, width=0.4, label='Train',color='mediumturquoise', ec='darkgray')
plt.bar(pear_AAR_names, pear_AAR_test, bottom=0, width=0.4, label='Test',color='lavender', ec='darkgray');
# plt.bar_label(react,labels=pear_AAR_test, label_type='edge', rotation=0, c='dimgray', fontsize=11)

plt.legend(fontsize=13)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.axhline(linewidth=1, color='darkgrey')
plt.ylabel("Pearson'r", fontsize=14)
plt.xlabel("Model type (AAIndex)", fontsize=14)
plt.text("Lr", -0.1, "Lr_Norm indicates that Normalised AAIndex values are used.")
plt.text("Lr", -0.17, "All the other models are trained using Normalized AAIndex")
plt.title("Best model fit is taken for each model random split (split=0.7, Filters=128, Convolution layer = 3xk17, Hidden layer NN=2x100, Batch size = 128)", loc='center',wrap='true',color="dimgrey",pad='17.0')
plt.text("Lr", -0.22, "values (Subtracted) except Lr model.")
# plt.savefig("Results/AA_Random.png")
plt.show()
```
