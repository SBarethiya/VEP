```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Comparison of max, avg and no pooling
# 1024 filters pool size 2 (501, ) --> (250,)
# data
# lr0.01 :3, lr 0.001 :3, 
# B32, B32, B32
pearsons_train = [0.46, 0.48, 0.50, 0.47, 0.43, 0.46, 0.52, 0.55, 0.54]
pearsons_test = [0.25, 0.33, 0.31, 0.33, 0.29, 0.32, 0.27, 0.29, 0.29]
r2_train = [0.17, 0.19, 0.18, 0.20, 0.17, 0.21, 0.26, 0.29, 0.28]
r2_test = [0.05, 0.10, 0.08, 0.11, 0.06, 0.09, 0.06, 0.06, 0.06]
names = ["Flatten_0.01", "Avg_0.01", "Max_0.01", "Flatten_0.001", "Avg_0.001", "Max_0.001", "Flatten_0.0001", "Avg_0.0001", "Max_0.0001"]

plt.figure(figsize=(8,5))
plt.ylim(-0.3, 1.0)
plt.grid(color='b', ls = '-.', lw = 0.25, alpha=0.5)
react = plt.bar(np.arange(1, len(names)+1), pearsons_train, width=0.4, label='Train',color='mediumturquoise', ec='darkgray')
plt.bar(np.arange(1, len(names)+1), pearsons_test, bottom=0, width=0.4, label='Test',color='lavender', ec='darkgray');
plt.bar_label(react,labels=names, label_type='center', rotation=90, c='dimgray', fontsize=11)
plt.legend(fontsize=13)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.axhline(linewidth=1, color='darkgrey')
plt.ylabel("Pearson'r", fontsize=14)
plt.xlabel("Pooling layer type", fontsize=14)
title = "For PyRosetta dataset, comparision of Max, average (Avg), and no pooling (Flatten) with comination of different learning rate (Batch size = 32, pooling size = 2, Filters = 1024, Split=70)."
plt.title(title, loc='center',wrap='true',color="dimgrey",pad='17.0')
plt.savefig("poolandlr_pyros_pearson_1024.png")
plt.show()

plt.cla()
plt.clf()
plt.figure(figsize=(8,5))
plt.ylim(-0.3, 1.0)
plt.grid(color='b', ls = '-.', lw = 0.25, alpha=0.5)
react = plt.bar(np.arange(1, len(names)+1), r2_train, width=0.4, label='Train',color='mediumturquoise', ec='darkgray')
plt.bar(np.arange(1, len(names)+1), r2_test, bottom=0, width=0.4, label='Test',color='lavender', ec='darkgray');
plt.bar_label(react,labels=names, label_type='center', rotation=90, c='dimgray', fontsize=11)
plt.legend(fontsize=13)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.axhline(linewidth=1, color='darkgrey')
plt.ylabel("r2", fontsize=14)
title = "For PyRosetta dataset, comparision of Max, average (Avg), and no pooling (Flatten) with comination of different learning rate (Batch size = 32, pooling size = 2, Filters = 1024, Split=70)."
plt.title(title, loc='center',wrap='true',color="dimgrey",pad='17.0')
plt.xlabel("Pooling layer type", fontsize=14)
plt.savefig("poolandlr_pyros_r2_1024.png")
```
