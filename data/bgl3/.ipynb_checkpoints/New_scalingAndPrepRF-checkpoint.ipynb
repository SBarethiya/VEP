```python
score = data["score"]
score
score_sc = []
for i in range(0, len(score),1):
    if (score[i]>50.0) or (score[i]<-50.0):
        dd = 50*np.tanh(score[i]/50)
        score_sc.append(dd)
    else:
        score_sc.append(score[i])
```

```python
fa_rep = data["fa_rep"]
fa_rep_sc = []
for i in range(0, len(fa_rep),1):
    if (fa_rep[i]>50.0) or (fa_rep[i]<-50.0):
        dd = 50*np.tanh(fa_rep[i]/50)
        fa_rep_sc.append(dd)
    else:
        fa_rep_sc.append(fa_rep[i])
```

```python
pro_close = data["pro_close"]
pro_close_sc = []
for i in range(0, len(pro_close),1):
    if (pro_close[i]>50.0) or (pro_close[i]<-50.0):
        dd = 50*np.tanh(pro_close[i]/50)
        pro_close_sc.append(dd)
    else:
        pro_close_sc.append(pro_close[i])
```

```python
data["score"] = score_sc
data = data.rename(columns={"score":"ddG"})
data["fa_rep"] = fa_rep_sc
data["pro_close"] = pro_close_sc
data
```
