```python
#Import all the needed packages
import csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#load the Rosetta data as dataframe
data = pd.read_csv("bgl3_Updated_woMute_12.csv")
data
```

```python
#Remove the index column (I forgot to do index false while writing the dataframe to csv)
data  = data.loc[:, ~data.columns.str.contains('^Unnamed')]
data
```

```python
#Remove deletions
data_paper =pd.read_csv("bgl3_original.tsv",sep='\t')
mutation = data_paper["variant"]
tot = len(mutation)
for i in range(0,tot,1):
    str1 = mutation[i]
    str2 = str1.split(',')
    if "*" in str(str2):
        print(i,mutation[i])
        mutation.pop(i)
        data_paper = data_paper.drop([i],axis=0)
        data = data.drop([i],axis=0)
data
```

```python
#After removing the deletions
data_paper = data_paper.reset_index()
data_paper  = data_paper.loc[:, ~data_paper.columns.str.contains('index')]
data_paper
data_paper.to_csv('bgl3_Sub.tsv',sep='\t',index=False)
data_paper
```

```python
#Remove mutation not necessary
data  = data.loc[:, ~data.columns.str.contains('mutation')]
data
```

```python
#Reset index after dropping rows
data = data.reset_index()
data  = data.loc[:, ~data.columns.str.contains('index')]
data
```

```python
#To check the range of data
for i in range(0,len(data.columns),1):
    x = data.columns[i]
    y = data[x]
    print(x,min(y),max(y))
```

```python
# Finding the corelateion between the ddG and function 
from scipy.stats import pearsonr
from scipy import stats

ddg = data['score']
function_score = data_paper['score']

corr, _ = pearsonr(ddg,function_score)
print('org p',corr)
res = stats.spearmanr(ddg,function_score)
print('org sp',res)
```

```python
dF_1 = []
dF_2 = []
# dF_3 = []
# dF_4 = []
# dF_5 = []
ddg_1 = []
ddg_2 = []
# ddg_3 = []
# ddg_4 = []
# ddg_5 = []
mutation = data_paper['variant']
for i in range(0, len(mutation),1):
    str1 = mutation[i]
    str2 = str1.split(',')
    if (ddg[i] < 5000.0):
        if len(str2) == 1:
            dF_1.append(function_score[i])
            ddg_1.append(ddg[i])
        elif len(str2) == 2:
            ddg_2.append(ddg[i])
            dF_2.append(function_score[i])
#         elif len(str2) == 3:
#             dF_3.append(function_score[i])
#             ddg_3.append(ddg[i])
#         elif len(str2) == 4:
#             ddg_4.append(ddg[i])
#             dF_4.append(function_score[i])
#         elif len(str2) == 5:
#             dF_5.append(function_score[i])
#             ddg_5.append(ddg[i])
print("Number of mutations N=1",len(ddg_1),len(dF_1))
print("Number of mutations N=2",len(ddg_2),len(dF_2))
# print("Number of mutations N=3",len(ddg_3),len(dF_3))
# print("Number of mutations N=4",len(ddg_4),len(dF_4))
# print("Number of mutations N=5",len(ddg_5),len(dF_5))

plt.scatter(ddg_2,dF_2,color='red',  label='2',s=1)
plt.scatter(ddg_1,dF_1,color='green',label='1',s=10)
# plt.scatter(ddg_3,dF_3,color='blue', label='3',s=1)
# plt.scatter(ddg_4,dF_4,color='black',label='4',s=1)
# plt.scatter(ddg_5,dF_5,color='pink', label='5',s=1)
plt.legend(loc='best')
plt.ylabel("dF")
plt.xlabel("ddG")
plt.show()
```

```python
# putting limit
score = data["score"]
score
score_sc = []
for i in range(0, len(score),1):
    if (score[i]>50.0) or (score[i]<-50.0):
        dd = 50*np.tanh(score[i]/50)
        score_sc.append(dd)
    else:
        score_sc.append(score[i])
```

```python
fa_rep = data["fa_rep"]
fa_rep_sc = []
for i in range(0, len(fa_rep),1):
    if (fa_rep[i]>50.0) or (fa_rep[i]<-50.0):
        dd = 50*np.tanh(fa_rep[i]/50)
        fa_rep_sc.append(dd)
    else:
        fa_rep_sc.append(fa_rep[i])
```

```python
pro_close = data["pro_close"]
pro_close_sc = []
for i in range(0, len(pro_close),1):
    if (pro_close[i]>50.0) or (pro_close[i]<-50.0):
        dd = 50*np.tanh(pro_close[i]/50)
        pro_close_sc.append(dd)
    else:
        pro_close_sc.append(pro_close[i])
```

```python
data["score"] = score_sc
data = data.rename(columns={"score":"ddG"})
data["fa_rep"] = fa_rep_sc
data["pro_close"] = pro_close_sc
data
```

```python
#To check the range of data
for i in range(0,len(data.columns),1):
    x = data.columns[i]
    y = data[x]
    print(x,min(y),max(y))
```

```python
data.to_csv('bgl3_ros_scaled.tsv',sep='\t',index=False)
```
