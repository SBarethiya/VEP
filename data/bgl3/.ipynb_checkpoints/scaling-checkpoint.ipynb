```python
score = data["score"]
score
```

```python
score_sc = []
for i in range(0, len(score),1):
    if (score[i]>50.0) or (score[i]<-50.0):
        dd = 50*np.tanh(score[i]/50)
        score_sc.append(dd)
        print(dd,score[i])
    else:
        score_sc.append(score[i])
        print(score[i])
```

```python
plt.
plt.hist(score,bins=500)
plt.ylabel('No. of Datapoints')
plt.xlabel('ddG')

plt.show()
plt.hist(score_sc,bins=500)
plt.ylabel('No. of Datapoints')
plt.xlabel('Scaled ddG')
plt.show()
```

```python
fa_rep = data["fa_rep"]
fa_rep_sc = []
for i in range(0, len(fa_rep),1):
    if (fa_rep[i]>50.0) or (fa_rep[i]<-50.0):
        dd = 50*np.tanh(fa_rep[i]/50)
        fa_rep_sc.append(dd)
        print(dd,fa_rep[i])
    else:
        fa_rep_sc.append(fa_rep[i])
        print(fa_rep[i])
```

```python
plt.hist(fa_rep,bins=500)
plt.ylabel('No. of Datapoints')
plt.xlabel('fa_rep')
plt.show()
plt.hist(fa_rep_sc,bins=500)
plt.ylabel('No. of Datapoints')
plt.xlabel('Scaled fa_rep')
plt.show()
```

```python
plt.grid(color='grey', linestyle='-', linewidth=0.2)
plt.hist(score_sc,color='green',label='ddG',bins=100)
plt.hist(fa_rep_sc,alpha = 0.5, lw=0.5,color='pink',label='Fa_rep',bins=100)
plt.ylabel('No. of Datapoints')
plt.xlabel('Scaled Scores')

plt.legend()
# plt.legend({'ddG','Fa_rep'})
plt.show()
```
